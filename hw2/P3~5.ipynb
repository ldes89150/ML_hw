```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import brentq
plt.style.use('ggplot')
%matplotlib inline
```

```python
N = np.linspace(0,500000)
def lhs(x):
    return x
def rhs(x):
    return (8/0.05**2)*np.log((4*(x)**10+4)/(0.05))
def f(x):
    return lhs(x)-rhs(x)
```

```python
plt.plot(N,lhs(N), label = "LHS")
plt.plot(N,rhs(N), label = "RHS")
plt.xlabel("N")
plt.legend()
plt.savefig("P3")
```

```python
brentq(f,400000,500000)
```

```python
N= np.linspace(2,10000,dtype=np.longdouble)
def m(N):
    return N**50
def a(N):
    return np.sqrt(8/N*np.log((4*m(2*N))/0.05))
def b(N):
    return np.sqrt(16/N*np.log((2*m(2*N))/0.05))
def c(N):
    return np.sqrt(2*np.log(2*N*m(N))/N)+np.sqrt(2/N*np.log(1/0.05))+1/N

def d(N):
    def _(epsilon):
        return epsilon-np.sqrt(1/N*(2*epsilon+np.log(6*m(2*N)/0.05)))    
    return brentq(_,0,10.0)
def d_array(N):
    r = []
    for n in N:
        try:
            r.append(d(n))
        except:
            r.append(np.NaN)
    return np.array(r)

def e(N):
    def _(epsilon):
        return epsilon-np.sqrt(1.0/2.0/N*(4*epsilon*(1+epsilon)+np.log(4.0)+100.0*np.log(N)-np.log(0.05)))
    return brentq(_,0.01,10.0)

def e_array(N):
    r = []
    for n in N:
        try:
            r.append(e(n))
        except:
            r.append(np.NaN)
    return np.array(r)
```

```python
N1= np.linspace(9000,10000)
N2= np.linspace(5,10)
plt.figure()
plt.subplot(2,1,1)
plt.plot(N,a(N), label = "Original VC bound")
plt.plot(N,b(N), label = "Variant VC bound")
plt.plot(N,c(N), label = "Rademacher Penalty Bound")
plt.plot(N,d_array(N), label = "Parrondo and Van den Broek")
plt.plot(N,e_array(N), label = "Devroye")
plt.legend()
plt.subplot(2,2,3)
plt.plot(N1,a(N1), label = "Original VC bound")
plt.plot(N1,b(N1), label = "Variant VC bound")
plt.plot(N1,c(N1), label = "Rademacher Penalty Bound")
plt.plot(N1,d_array(N1), label = "Parrondo and Van den Broek")
plt.plot(N1,e_array(N1), label = "Devroye")
plt.subplot(2,2,4)
plt.plot(N2,a(N2), label = "Original VC bound")
plt.plot(N2,b(N2), label = "Variant VC bound")
plt.plot(N2,c(N2), label = "Rademacher Penalty Bound")
plt.plot(N2,d_array(N2), label = "Parrondo and Van den Broek")
plt.plot(N2,e_array(N2), label = "Devroye")
plt.savefig("P4")
```

```python
def v(N):
    return a(N)-0.05
brentq(v,400000,500000)
```
