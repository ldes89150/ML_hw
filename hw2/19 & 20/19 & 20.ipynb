```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('ggplot')
from IPython.display import display, Math, Latex
%matplotlib inline
```

```python
train_data = pd.read_csv("hw2_train.dat", sep= ' ' ,names = ["strip"]+["x{0}".format(i) for i in  range(9)] + ["s"])
train_data =train_data.drop("strip", 1)
xs = [np.array(train_data["x{0}".format(i)]) for i in range(9)]
y = np.array(train_data["s"])
train_data
```

```python
def h(x, theta ,s):
    return s*np.sign(x-theta)

def get_hyposis_set(x):
    x = np.sort(x)
    mediate= [(x[:-1]+x[1:])/2, np.array([1]), (x[:-1]+x[1:])/2, np.array([1])]
    theta_vector = np.concatenate(mediate)
    n = len(x)
    s_vector = np.concatenate([[1 for i in range(n)],[-1 for i in range(n)]])
    return theta_vector, s_vector

def E(x, y, theta, s):
    result = h(x, theta, s) == y
    return 1- np.average(result)
```

```python
best_s = []
best_theta = []
best_E_in = []
for i in xrange(9):
    x = xs[i]
    theta_vector, s_vector = get_hyposis_set(x)
    best_round_E_in = 1
    best_round_theta = None
    best_round_s = None
    for j in xrange(2*len(x)):
        theta = theta_vector[j]
        s = s_vector[j]
        round_E_in = E(x, y, theta, s)
        if round_E_in < best_round_E_in:
            best_round_theta = theta
            best_round_s = s
            best_round_E_in = round_E_in
    best_s.append(best_round_s)
    best_theta.append(best_round_theta)
    best_E_in.append(best_round_E_in)
optimal_i = np.argmin(best_E_in)
optimal_theta = best_theta[optimal_i]
optimal_s = best_s[optimal_i]
optimal_E_in = best_E_in[optimal_i]
```

```python
display(Latex(r"$$i = {0}\\ \theta = {1}\\ s = {2}\\ E_{{in}} = {3}$$".format(optimal_i+1, optimal_theta, optimal_s, optimal_E_in)))
```

```python
test_data = pd.read_csv("hw2_test.dat", sep= ' ' ,names = ["strip"]+["x{0}".format(i) for i in  range(9)] + ["s"])
test_data =test_data.drop("strip", 1)
xs_test = [np.array(test_data["x{0}".format(i)]) for i in range(9)]
y_test = np.array(test_data["s"])
test_data
```

```python
E_outs = []
for i in xrange(9):
    x_test = xs_test[i]
    E_outs.append(E(x_test, y_test, optimal_theta, optimal_s))
```

```python
display(Latex(r"$$E_{{out}} = {0}$$".format(np.min(E_outs))))
```
